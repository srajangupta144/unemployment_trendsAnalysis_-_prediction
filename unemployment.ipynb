```python
import pandas as pd

df= pd.read_csv('Unemployment1.csv')
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df[df.duplicated()]
```

```python
columns_to_remove = [f'Unemployed_{year}' for year in range(2000, 2023)]

df = df.drop(columns=columns_to_remove)

print(df.columns)
```

```python
labor_force_cols = [col for col in df.columns if 'Civilian_labor_force' in col]
employed_cols = [col for col in df.columns if 'Employed' in col and 'Med' not in col]
unemployment_rate_cols = [col for col in df.columns if 'Unemployment_rate' in col]

df_labor_force = df.melt(
    id_vars=['FIPS_Code', 'State', 'Area_Name',
             'Rural_Urban_Continuum_Code_2013',
             'Urban_Influence_Code_2013',
             'Metro_2013'],
    value_vars=labor_force_cols,
    var_name='Year',
    value_name='Civilian_Labor_Force'
)

df_employed = df.melt(
    id_vars=['FIPS_Code', 'State', 'Area_Name',
             'Rural_Urban_Continuum_Code_2013',
             'Urban_Influence_Code_2013',
             'Metro_2013'],
    value_vars=employed_cols,
    var_name='Year',
    value_name='Employed'
)

df_unemployment_rate = df.melt(
    id_vars=['FIPS_Code', 'State', 'Area_Name',
             'Rural_Urban_Continuum_Code_2013',
             'Urban_Influence_Code_2013',
             'Metro_2013'],
    value_vars=unemployment_rate_cols,
    var_name='Year',
    value_name='Unemployment_Rate'
)

df_labor_force['Year'] = df_labor_force['Year'].str.extract('(\d+)$')
df_employed['Year'] = df_employed['Year'].str.extract('(\d+)$')
df_unemployment_rate['Year'] = df_unemployment_rate['Year'].str.extract('(\d+)$')

df_combined = df_labor_force.merge(df_employed, on=['FIPS_Code', 'State', 'Area_Name',
                                                    'Rural_Urban_Continuum_Code_2013',
                                                    'Urban_Influence_Code_2013',
                                                    'Metro_2013', 'Year'])
df_combined = df_combined.merge(df_unemployment_rate, on=['FIPS_Code', 'State', 'Area_Name',
                                                          'Rural_Urban_Continuum_Code_2013',
                                                          'Urban_Influence_Code_2013',
                                                          'Metro_2013', 'Year'])

# Display the reshaped data
print(df_combined.head())
```

```python
df_combined.to_csv('reshaped_unemployment_data.csv', index=False)

from google.colab import files
files.download('reshaped_unemployment_data.csv')

```
